```python
# Define income band boundaries
# Converting Estimated Income into catagorical data
bins = [0, 100000, 300000, float('inf')]
labels = ['low', 'mid', 'High']

df['Income Band'] = pd.cut(df['Estimated Income'], bins=bins , labels=labels, right=False)
```

```python
df['Income Band'].value_counts().plot(kind="bar")
```

```python
# Examine the distribution of unique catagoried in categorical col
categorical_cols = df[["Amount of Credit Cards","Nationality","Fee Structure","Loyalty Classification","Properties Owned","Risk Weighting","Occupation","Income Band", "BRId", "IAId"]].columns

for col in categorical_cols:
    print(f"Value count for '{col}':")
    display(df[col].value_counts())
```

## Univariate analysis

```python
# ploting the distribution of unique catagoried in categorical col

for i, col in enumerate(categorical_cols, start=1):
    
    # --- special case: only topâ€‘5 occupations ---
    if col == "Occupation":
        top_vals = df[col].value_counts().nlargest(10).index          # top 5 labels
        data     = df[df[col].isin(top_vals)]                         # keep only those rows
        order    = top_vals                                           # bar order
        title    = f"Topâ€¯10 {col}s"
    else:
        data     = df                                                 # full data
        order    = df[col].value_counts().index                       # all labels, mostâ€‘common first
        title    = f"Distribution of {col}"
    
    # --- plotting ---
    plt.figure(i, figsize=(6, 4))
    sns.countplot(data=data, x=col, order=order)
    plt.title(title)
    plt.xticks(rotation=45, ha="right")
    plt.tight_layout()
    plt.show()
```

## Bivariate analysis

```python
# Create subplot grid â€” adjust rows/cols as needed
fig, axes = plt.subplots(nrows=4, ncols=2, figsize=(15, 18))
axes = axes.flatten()  # make it easier to loop

for i, col in enumerate(categorical_cols):
    
    ax = axes[i]
    
    # Special handling for Occupation (Top 10)
    if col == "Occupation":
        top_vals = df[col].value_counts().nlargest(10).index
        data     = df[df[col].isin(top_vals)]
        order    = top_vals
        title    = f"Top 10 {col}s"
    else:
        data     = df
        order    = df[col].value_counts().index
        title    = f"{col}"
    
    sns.countplot(data=data, x=col, order=order, hue="Nationality", ax=ax)
    ax.set_title(title)
    ax.tick_params(axis='x', rotation=45)

# Turn off any unused subplot spaces
for j in range(len(categorical_cols), len(axes)):
    fig.delaxes(axes[j])

plt.tight_layout()
plt.show()

    
```

## Numerical Analysis

```python
numerical_cols = ['Fee Structure', 'Age','Estimated Income', 'Superannuation Savings', 'Credit Card Balance', 'Bank Loans', 'Bank Deposits', 'Checking Accounts', 'Saving Accounts', 'Foreign Currency Account', 'Business Lending']

# Univariate analysis and visualisation
plt.figure(figsize=(15, 10))
for i,col in enumerate(numerical_cols):
    plt.subplot(4,3,i+1)
    sns.histplot(df[col],kde=True)
    plt.title(col)
plt.tight_layout()
plt.show()
```

## Heatmaps

```python
numerical_cols = ['Age', 'Estimated Income', 'Superannuation Savings', 'Credit Card Balance',
                  'Bank Loans', 'Bank Deposits', 'Checking Accounts', 'Saving Accounts',
                  'Foreign Currency Account', 'Business Lending', 'Properties Owned']
correlation_matrix = df[numerical_cols].corr()

plt.figure(figsize=(8,8))
sns.heatmap(correlation_matrix, annot=True, cmap='crest', fmt=".2f")
plt.title('Correlation Matrix of Numerical Features')
plt.show()
```
