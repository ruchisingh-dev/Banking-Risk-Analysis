```python
# Define income band boundaries
# Converting Estimated Income into catagorical data
bins = [0, 100000, 300000, float('inf')]
labels = ['low', 'mid', 'High']

df['Income Band'] = pd.cut(df['Estimated Income'], bins=bins , labels=labels, right=False)
```

```python
df['Income Band'].value_counts().plot(kind="bar")
```

```python
# Examine the distribution of unique catagoried in categorical col
categorical_cols = df[["Amount of Credit Cards","Nationality","Fee Structure","Loyalty Classification","Properties Owned","Risk Weighting","Occupation","Income Band", "BRId", "IAId"]].columns

for col in categorical_cols:
    print(f"Value count for '{col}':")
    display(df[col].value_counts())
```

## Univariate analysis

```python
# ploting the distribution of unique catagoried in categorical col

for i, col in enumerate(categorical_cols, start=1):
    
    # --- special case: only topâ€‘5 occupations ---
    if col == "Occupation":
        top_vals = df[col].value_counts().nlargest(10).index          # top 5 labels
        data     = df[df[col].isin(top_vals)]                         # keep only those rows
        order    = top_vals                                           # bar order
        title    = f"Topâ€¯10 {col}s"
    else:
        data     = df                                                 # full data
        order    = df[col].value_counts().index                       # all labels, mostâ€‘common first
        title    = f"Distribution of {col}"
    
    # --- plotting ---
    plt.figure(i, figsize=(6, 4))
    sns.countplot(data=data, x=col, order=order)
    plt.title(title)
    plt.xticks(rotation=45, ha="right")
    plt.tight_layout()
    plt.show()
```

## Bivariate analysis

```python
for i, col in enumerate(categorical_cols, start=1):
    
    # --- special case: only topâ€‘5 occupations ---
    if col == "Occupation":
        top_vals = df[col].value_counts().nlargest(10).index          # top 5 labels
        data     = df[df[col].isin(top_vals)]                         # keep only those rows
        order    = top_vals                                           # bar order
        title    = f"Topâ€¯10 {col}s"
    else:
        data     = df                                                 # full data
        order    = df[col].value_counts().index                       # all labels, mostâ€‘common first
        title    = f"Distribution of {col}"
    
    # --- plotting ---
    plt.figure(i, figsize=(6, 4))
    sns.countplot(data=data, x=col, order=order, hue="Nationality")
    plt.title(title)
    plt.xticks(rotation=45, ha="right")
    plt.tight_layout()
    plt.show()
    
```
